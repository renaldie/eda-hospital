```python
from paddleocr import PaddleOCRVL
import os
import glob

for folder in os.listdir("./assets/input"):
    print(f"In folder: {folder}")
    for file in os.listdir(f"./assets/input/{folder}"):
        # Construct the expected output filename
        base_name = os.path.splitext(file)[0]
        
        # Check if output file already exists in any format:
        # 1. Paginated format: base_name_0.md, base_name_1.md, etc.
        # 2. Combined format: base_name.md (without page numbers)
        output_folder = f"./assets/output-combined/{folder}"
        
        # Check for paginated files
        paginated_exists = os.path.exists(os.path.join(output_folder, f"{base_name}_0.md"))
        
        # Check for combined file (without _0, _1 suffix)
        combined_exists = os.path.exists(os.path.join(output_folder, f"{base_name}.md"))
        
        # Check for any numbered variants using glob
        any_page_exists = len(glob.glob(os.path.join(output_folder, f"{base_name}_*.md"))) > 0
        
        if paginated_exists or combined_exists or any_page_exists:
            print(f"Skipping {file} - already converted (paginated or combined)")
            continue

        print(f"Converting {file}")
        pipeline = PaddleOCRVL(
            use_doc_orientation_classify=True,
            use_doc_unwarping=True, 
            use_layout_detection=True,
            use_chart_recognition=True,
        )
        output = pipeline.predict(
            input=f"./assets/input/{folder}/{file}",
            use_doc_orientation_classify=True,
            use_doc_unwarping=True,
            use_layout_detection=True,
            use_chart_recognition=True,
        )
        for res in output:
            res.save_to_json(save_path=output_folder)
            res.save_to_markdown(save_path=output_folder)
```

```python
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI
import os
import re
from langchain_core.messages import SystemMessage, HumanMessage

load_dotenv(override=True)
GITHUB_TOKEN = os.environ.get("GITHUB_TOKEN")
llm = ChatOpenAI(
    model="openai/gpt-4.1-mini",
    temperature=0,
    base_url="https://models.github.ai/inference",
    api_key=GITHUB_TOKEN,
)


def ch_to_eng(text: str) -> str:
    """Translate Chinese to English"""
    system = SystemMessage(
        content="You are a Mandarin to English translator. Only translate without adding or removing anything"
    )
    human = HumanMessage(content=text)
    resp = llm.invoke(input=[system, human])
    return resp.content


def extract_document_id(filename: str) -> str:
    """
    Extract document ID from filename (e.g., HA-7-0041 from the filename)
    Pattern: HA-X-XXXX where X are numbers
    """
    match = re.match(r'^(HA-\d+-\d+)', filename)
    if match:
        return match.group(1)
    return None


for folder in os.listdir("./assets/output"):
    print(f"\nIn folder: {folder}")
    for file in os.listdir(f"./assets/output/{folder}"):
        if file.endswith(".md") and not file.endswith("-ENG.md"):
            print(f"\nProcessing: {file}")
            
            # Extract document ID from the Chinese file
            doc_id = extract_document_id(file)
            
            if not doc_id:
                print(f"  âš ï¸  Warning: Could not extract document ID from {file}, skipping")
                continue
            
            # Check if any English translation with the same document ID already exists
            existing_eng_files = [
                f for f in os.listdir(f"./assets/output/{folder}")
                if f.startswith(doc_id) and f.endswith("-ENG.md")
            ]
            
            if existing_eng_files:
                print(f"  Skipping - already translated (found: {existing_eng_files[0]})")
                continue

            # Read the content
            with open(f"./assets/output/{folder}/{file}", encoding='utf-8') as f:
                content = f.read()

            # Get base name without extension and page number
            base_name = os.path.splitext(file)[0]
            base_name_no_page = re.sub(r'_\d+$', '', base_name)
            
            # Translate
            print("Translating title...")
            translated_title = ch_to_eng(base_name_no_page)

            print("Translating content...")
            translated_content = ch_to_eng(content)

            # New naming format: doc_id_TranslatedText-ENG.md
            # We keep the translated title from the filename pattern
            output_filename = f"{translated_title}-ENG.md"
            
            with open(f"./assets/output/{folder}/{output_filename}", "w", encoding='utf-8') as f:
                f.write(translated_content)
            
            print(f"  âœ“ Created: {output_filename}")

print(f"\n\nâœ… Translation complete!")
```
