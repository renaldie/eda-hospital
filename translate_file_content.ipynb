```python
import os
from langchain_ollama import ChatOllama
from langchain_core.messages import SystemMessage, HumanMessage
```

```python
llm = ChatOllama(model="qwen3:1.7b",
                 temperature=0,
                 max_tokens=128)
```

```python
def ch_to_eng(text: str) -> str:
    system = SystemMessage(
        content="""
        You are a Mandarin to English translator. 
        Only translate the following markdown content without adding or removing anything.
        Retain all of the markdown structure
        """
    )
    human = HumanMessage(content=text)
    resp = llm.invoke(input=[system, human])
    return resp.content
```

```python
for folder in os.listdir("./assets/output-en"):
    for filename in os.listdir(f"./assets/output-en/{folder}"):
        if filename.endswith(".md"):
            filepath = f"./assets/output-en/{folder}/{filename}"
            
            with open(filepath, "r", encoding="utf-8") as f:
                content = f.read()
            
            translated_content = ch_to_eng(content)
            
            with open(filepath, "w", encoding="utf-8") as f:
                f.write(translated_content)
```
